```python
"visualizing categorical plot"
#The default representation of the data in catplot() uses a scatterplot
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.catplot(data=tips,x='day',y='total_bill' )
```

```python
#the jitter parameter controls the magnitude of jitter
sns.catplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
#prevent from overlapping (swarm plot)
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
#categorical plotting function handles the hue semantic
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='swarm')
```

```python
#order parameter(to display multiple categorical plpots in the same figure)
sns.catplot(data=tips,x='smoker',y='tip',order=["No","Yes"])
```

```python
#categorical varaiable on the vertical axis
sns.catplot(data=tips,x='total_bill',y='day',hue='time',kind='swarm')
```

```python
#Boxplots
sns.catplot(data=tips,x='day',y='total_bill',kind='box')
```

```python
#adding hue sematic
sns.catplot(data=tips,x='day',y='total_bill',hue='smoker',kind='box')
```

```python
#violin plt
sns.catplot(data=tips,x='total_bill',y='day',hue='sex',kind='violin')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
data.head(10)
```

```python
data.tail()
```

```python
data.info()
```

```python
data.isna()
```

```python
from numpy import nan
data['Glucose'] = data['Glucose'].replace(0,np.nan)
data['BloodPressure'] = data['BloodPressure'].replace(0,np.nan)
data['SkinThickness'] = data['SkinThickness'].replace(0,np.nan)
data['Insulin'] = data['Insulin'].replace(0,np.nan)
data['BMI'] = data['BMI'].replace(0,np.nan)
```

```python
data.head()
```

```python
#count of NaN values in the dataset
print(data.isnull().sum())
```

```python
data.median()
```

```python
data.fillna(data.median(),inplace=True)
```

```python
#checking the missing values
print(data.isnull().sum())
```

```python
plt.figure(figsize=(20,19))
plt.subplot(4,4,1)
sns.boxplot(data['Pregnancies'])

plt.subplot(4,4,2)
sns.boxplot(data['Glucose'])

plt.subplot(4,4,3)
sns.boxplot(data['BloodPressure'])

plt.subplot(4,4,4)
sns.boxplot(data['SkinThickness'])

plt.subplot(4,4,5)
sns.boxplot(data['Insulin'])

plt.subplot(4,4,6)
sns.boxplot(data['BMI'])

plt.subplot(4,4,7)
sns.boxplot(data['DiabetesPedigreeFunction'])

plt.subplot(4,4,8)
sns.boxplot(data['Age'])
```

```python
# Filter out outliers using the 10th and 90th percentiles
data['SkinThickness'] = data['SkinThickness'][
    (data['SkinThickness'] > data['SkinThickness'].quantile(0.10)) &
    (data['SkinThickness'] < data['SkinThickness'].quantile(0.90))
]

data['Insulin'] = data['Insulin'][
    (data['Insulin'] > data['Insulin'].quantile(0.10)) &
    (data['Insulin'] < data['Insulin'].quantile(0.90))
]

# Plot boxplot for ShiftIncness
plt.figure(figsize=(8,6))
sns.boxplot(data['SkinThickness'])
plt.show()

# Plot boxplot for Insulin
plt.figure(figsize=(8,6))
sns.boxplot(data['Insulin'])
plt.show()
```
